```python
import pandas as pd
%pylab inline
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'Questions for dataset.csv'
questions = pd.read_csv(path)
```

```python
destinations = questions["What is your next holiday destination"].tolist()
destinations
```

```python
plt.hist(questions['How long do you spend on your phone in a day in average'], bins=25)
plt.xlabel('hours')
plt.ylabel('Number of people')
plt.title('Daily Hours Spending on Phone')
```

```python
hours_groupings = questions.groupby('What is your favourite crisps flavour')
hours_groupings['How long do you spend on your phone in a day in average'].unique()
```

```python
hours_groupings['How long do you spend on your phone in a day in average'].describe()
```
